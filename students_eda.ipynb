```python
students=data.drop(['race/ethnicity','parental level of education'],axis=1)
students
```

```python
correlation=students.corr(numeric_only=True)
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(students)
```

```python
sns.relplot(x='math score',y='reading score',hue='lunch',data=students)
```

```python
sns.distplot(students['math score'])
```

```python
sns.distplot(students['reading score'])
```

```python
sns.distplot(students['writing score'],bins=5)
```
